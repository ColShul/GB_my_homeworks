Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.
Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.


```python
#Сoздаем массив t
#Округляем его для "красивого" размещения цифр на оси Х
t= np.round(np.linspace(0, 10, 51), decimals = 1)
t
```

```python
# С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.
x= np.linspace(-3, 3, 51)
x
```

```python
# Создайте массивы y1, y2, y3, y4 по следующим формулам:
# y1 = x**2
# y2 = 2 * x + 0.5
# y3 = -3 * x - 1.5
# y4 = sin(x)

y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
y1, y2, y3, y4
```

```python
# Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure
# с названием fig и массив объектов Axes под названием ax,причем так, чтобы было 4 отдельных графика в сетке,
# состоящей из двух строк и двух столбцов.
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()


# Чертим графики
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)

# Устанавливаем размер пространства для размещения графиков в дюймах
fig.set_size_inches(8, 6)

# Устанавливаем вертикальные и горизонтальные зазоры между графиками равными 0.3.
fig.subplots_adjust(wspace=0.3, hspace=0.3)

# Присваиваем названия графикам
ax1.set_title("График y1")
ax2.set_title("График y2")
ax3.set_title("График y3")
ax4.set_title("График y4")

# Для графика в левом верхнем углу задаем границы по оси x от -5 до 5
ax1.set_xlim([-5, 5])
```

```python
# Создать одномерный массив а, состоящий из последовательности целых чисел от 0 до 12.
# Поменять форму этого массива, чтобы получилась матрица A (двумерный массив Numpy), 
# состоящая из 4 строк и 3 столбцов. Получить матрицу At путем транспонирования матрицы A. 
# Получить матрицу B, умножив матрицу A на матрицу At с помощью матричного умножения. Какой размер имеет матрица B? 
# Получится ли вычислить обратную матрицу для матрицы B и почему?

a = np.array(np.linspace(0, 12, 12), dtype = int)
print ("Array a: \n", a)  
A = a.reshape(4, 3)
print ("Matrix A: \n", A)  
A_t = A.T
print ("Matrix A_t - after transporting matrix A: \n", A_t)  
```

```python
B = np.dot(A, A_t)
print ("Matrix B: \n", B) 
```

```python
B_i=np.linalg.inv(B)
B_i
# Вычислить обратную матрицу для матрицы B не получится, т.к. это сингулярная матрица
```

```python
# Инициализируйте генератор случайных числе с помощью объекта seed, равного 42
np.random.seed(42)

# Создайте одномерный массив c, составленный из последовательности 16-ти случайных 
# равномерно распределенных целых чисел от 0 до 16 невключительно. 
np.random.seed(42)
c = np.random.randint(0, 16, 16)
print ("Array c: \n", c)  
```

```python
# Поменяйте его форму так, чтобы получилась квадратная матрица C. 
C=c.reshape(4,4)
print ("Matrix C 4x4: \n", C) 

# Получите матрицу D, поэлементно прибавив матрицу B из предыдущего вопроса к матрице C, умноженной на 10. 
D = B + C*10
print ("Matrix D: \n", D) 

# Вычислите определитель, ранг и обратную матрицу D_inv для D.
print ("Определитель матрицы  D: ", np.linalg.det(D))
print ("Ранг матрицы  D: ", np.linalg.matrix_rank(D))
D_inv = np.linalg.inv(D)
print ("Обратная матрица для матрицы D: \n", D_inv) 
```

Приравняйте к нулю отрицательные числа в матрице D_inv, а положительные - к единице.
Убедитесь, что в матрице D_inv остались только нули и единицы.
С помощью функции numpy.where, используя матрицу D_inv в качестве маски,
а матрицы B и C - в качестве источников данных, получите матрицу E размером 4x4.
Элементы матрицы E, для которых соответствующий элемент матрицы D_inv равен 1,
должны быть равны соответствующему элементу матрицы B, а элементы матрицы E, 
для которых соответствующий элемент матрицы D_inv равен 0, должны быть равны соответствующему элементу матрицы C.


```python
D_inv = np.where(D_inv > 0, 1, 0)
print ("New matrix D_inv: \n", D_inv) 
E = np.where(D_inv == 1, B, C)
print ("Matrix B: \n", B) 
print ("Matrix C: \n", C) 
print ("Matrix E: \n", E) 
```
